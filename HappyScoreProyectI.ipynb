```python
import pandas as pd

data = pd.read_csv('happyscore_income.csv')

print (data)
```

```python
happy = data ['happyScore']
income = data ['avg_income']

print(happy)
```

```python
import matplotlib.pyplot as plt

plt.xlabel('income')
plt.ylabel('happy score')
plt.scatter(income, happy)
```

```python
#sorting
import numpy as np

data.sort_values('avg_income', inplace=True)

data

```

```python
#filtering

richest = data [data['avg_income'] > 15000]

richest
```

```python
#lowest income in the rich set 
richest .iloc [0]
```

```python
#highest income in the rich set 
richest .iloc [-1]
```

```python
#statistics
import numpy as np

rich_mean = np.mean(richest['avg_income'])

all_mean = np.mean (data['avg_income'])

print(all_mean, rich_mean)
```
